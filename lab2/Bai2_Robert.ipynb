```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = Image.open("coca.jpg")
img = img.convert("L")
img = np.array(img)
img = img.astype(np.float64)

img_final = np.zeros((len(img), len(img[0])))
gx = np.zeros((len(img), len(img[0])))
gy = np.zeros((len(img), len(img[0])))

for i in range(1, len(img) - 1):  # Height
    for j in range(1, len(img[0]) - 1):  # Width
        gx[i][j] = img[i + 1][j + 1] - img[i][j]
        gy[i][j] = img[i][j + 1] - img[i + 1][j]
        img_final[i][j] = min(255, np.sqrt(gx[i][j] ** 2 + gy[i][j] ** 2))

plt.figure(figsize=(12, 8))
plt.suptitle('PHƯƠNG PHÁP PHÁT HIỆN BIÊN BẰNG TOÁN TỬ ROBERT', fontsize=16)

plt.subplot(221)
plt.imshow(img, cmap='gray')
plt.title("Original")

plt.subplot(222)
plt.imshow(gx, cmap='gray', vmin=np.min(gx), vmax=np.max(gx))
plt.title("Vertical Robert")

plt.subplot(223)
plt.imshow(gy, cmap='gray', vmin=np.min(gy), vmax=np.max(gy))
plt.title("Horizontal Robert")

plt.subplot(224)
plt.imshow(img_final, cmap='gray', vmin=0, vmax=255)
plt.title("Robert")

plt.show()
```
